#### 1. Cálculo del Índice de Madurez

Este código ayuda a determinar si un fruto está listo para la cosecha basándose en la relación entre sólidos solubles (°Brix) y acidez titulable.

```
# Datos de entrada
brix = 12.5
acidez = 0.8

# Cálculo del índice
indice_madurez = brix / acidez

print(f"El índice de madurez es: {indice_madurez:.2f}")

if indice_madurez > 15:
    print("Estado: Fruto óptimo para cosecha/procesamiento.")
else:
    print("Estado: Aún requiere tiempo de maduración.")
```

#### 2. Balance de Masa Sencillo (Mezclado)

Imagina que quieres estandarizar el contenido de grasa en una mezcla láctea.

```
# Queremos saber cuánta leche descremada (0.5%) y crema (30%)
# se necesita para 100kg de leche al 3% de grasa.

def balance_grasa(total_deseado, porcentaje_objetivo):
    # Ecuación simple de balance
    crema = (total_deseado * porcentaje_objetivo) / 30
    leche_desc = total_deseado - crema
    return leche_desc, crema

desc, cr = balance_grasa(100, 3)
print(f"Para 100kg al 3%, usa {desc:.2f}kg de leche desc. y {cr:.2f}kg de crema.")
```

#### 3. Conversión de Unidades de Temperatura

Muy útil para monitorear autoclaves o cuartos fríos que usan diferentes escalas.

```
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

temperatura_c = 4.5  # Temperatura de refrigeración
temp_f = celsius_a_fahrenheit(temperatura_c)

print(f"La temperatura del cuarto frío es de {temperatura_c}°C ({temp_f}°F)")
```
